```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import pickle

# Load your pre-trained Random Forest model from the pickle file
with open('supply_model_1.pkl', 'rb') as f:
    rf_classifier = pickle.load(f)

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Late Delivery Prediction Dashboard"),
    html.Label("Days for shipment"),
    dcc.Input(id='Days for shipment', type='text', value=''),
    html.Label("Latitude"),
    dcc.Input(id='Latitude', type='number', value=''),
    html.Label("Longitude"),
    dcc.Input(id='Longitude', type='number', value=''),
    html.Label("Order City"),
    dcc.Input(id='Order City', type='text', value=''),
    html.Button(id='submit-button', n_clicks=0, children='Predict'),
    html.Div(id='output'),
dcc.Dropdown(
        id='Shipping Mode',
        options=[
            {'label': 'Standard Class', 'value': 'Standard Class'},
            {'label': 'First Class', 'value': 'First Class'},
            {'label': 'Second Class', 'value': 'Second Class'},
            {'label': 'Same Day', 'value': 'Same Day'}
        ],
        value='Shipping Mode',
        clearable=False
    ),
])

# Define callback to update the output
@app.callback(
    Output('output', 'children'),
    [Input('submit-button', 'n_clicks')],
    [dash.dependencies.State('Days for shipment', 'value'),
     dash.dependencies.State('Latitude', 'value'),
     dash.dependencies.State('Longitude', 'value'),
     dash.dependencies.State('Order City', 'value'),
     dash.dependencies.State('Shipping Mode', 'value')]
)
def update_output(n_clicks, feature1, feature2, feature3, feature4, feature5):
    if n_clicks > 0:
        # Make prediction using the loaded model
        prediction = rf_classifier.predict([[feature1, feature2, feature3, feature4, feature5]])[0]
        return f'Late delivery risk prediction: {prediction}'

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
